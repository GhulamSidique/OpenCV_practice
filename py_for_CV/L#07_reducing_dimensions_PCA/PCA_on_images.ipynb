### In tthis project we will apply PCA oto reduce the dimensions of images and reproduce and show the image

```python
# import the libraries
import cv2 as cv
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# load the image
dim_image=cv.cvtColor(cv.imread("E:/computer vision/py_for_CV/L#07_reducing_dimensions_PCA/image3.jpg"), cv.COLOR_BGR2RGB)
dim_image.shape
```

```python
# show the original image
plt.imshow(dim_image)
```

```python
# spliting the axes into three variables
r, g, b=cv.split(dim_image)

# dividing above three vars by 255
r, g,b=r/255, g/255, b/255
```

```python
# now visualizing above three vars separately
plt.imshow(r)
```

```python
plt.imshow(b)
```

```python
plt.imshow(g)
```

```python
# we will keep number of rows the same and will reduce the dimwnsion of columns
pca_comp=40 # we can increase or decrease this number according to the quality of image
pca_r=PCA(n_components=pca_comp)
pca_g=PCA(n_components=pca_comp)
pca_b=PCA(n_components=pca_comp)

# applying pca on above three vars
reduced_r=pca_r.fit_transform(r)
reduced_g=pca_g.fit_transform(g)
reduced_b=pca_b.fit_transform(b)
```

```python
# combining these reduced results into a single numpy array
red_combined=np.array([reduced_r, reduced_g, reduced_b])
# print(red_combined.shape)

# reshape above array
red_combined.reshape(1336, 40, 3).shape
```

```python
# noww applying the inverse on above reduced vars to reconstruct the original image with the reduced dimensions
rec_r=pca_r.inverse_transform(reduced_r)
rec_g=pca_g.inverse_transform(reduced_g)
rec_b=pca_b.inverse_transform(reduced_b)
```

```python
# now merge these results
merged_vars=(cv.merge((rec_r, rec_g, rec_b)))

# plot the reconstructed image
plt.imshow(merged_vars)
```

## Results:
### From above figure it is clear that the original image with (1336, 2546, 3) is same as the new image with reduced dimension that is 40.
### Therefore, the PCA algo worked here
